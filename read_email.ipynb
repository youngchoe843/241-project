```python
import smtplib
import time
import imaplib
import email
import re
from email.iterators import body_line_iterator

# -------------------------------------------------
#
# Utility to read email from Gmail Using Python
#
# ------------------------------------------------

# def read_email_from_gmail(FROM_EMAIL, FROM_PWD, SMTP_SERVER):
#     try:
mail = imaplib.IMAP4_SSL('imap.gmail.com')
mail.login('sarah.anderson1124@gmail.com','asdfjkl123')
mail.select('inbox')

type, data = mail.search(None, 'ALL')
mail_ids = data[0]
id_list = mail_ids.split()   
first_email_id = int(id_list[0])
latest_email_id = int(id_list[-1])
email_subject = list()
received_from = list()
received_date = list()
sent_date = list()
messages = list()

for i in range(latest_email_id,first_email_id, -1):
# #         for i in id_list:
    typ, data = mail.fetch(i, '(RFC822)' )
# typ, data = mail.fetch(300, '(RFC822)' )

    for response_part in data:
        if isinstance(response_part, tuple):
            msg = email.message_from_string(response_part[1])
            print msg['subject']
            email_subject.append(msg['subject'])
            received_from.append(msg['from'])
            received_date.append(msg['Date'])
            sent_date.append(msg['Received'])
            messages.append(msg.get_payload(None, True))

            text = ''
            for line in body_line_iterator(msg):
                text += line
    #             print line
    #             try:
    #                 a = re.search('\$', line).start()
    #                 b = re.search('0', line).end()
    #                 price[k] = line[a:b]
    #             except:
    #                 continue
            messages.append(text)

# return msg
```

```python
import pandas as pd
import numpy as np
# out_df = pd.DataFrame(np.reshape([email_subject,received_date, received_from, sent_date], (-1,4)))
out_df = pd.DataFrame([email_subject,received_date, received_from, sent_date, messages])
out_df = out_df.transpose()
out_df.columns = ['subject','date','from','sent_date','raw_text']
```

```python
import re
def filter_email(x):
    try:
        start = re.search('<', x).start()
        end = re.search('>', x).end()
        return x[start+1:end-1]
    except:
        return x
    
```

```python
# Exclude the failed delivery
clean_df = out_df.loc[out_df['subject'].str.contains('Delivery')==False, :]
clean_df = clean_df.loc[clean_df['subject'].str.contains('Undelivered')==False, :]
clean_df = clean_df.loc[clean_df['subject'].str.contains('Undeliverable')==False, :]

# Capture the date
clean_df['sent_date'] = map(lambda x: x[-32:-12], clean_df['sent_date'])
clean_df['date'] = map(lambda x: x[5:-6], clean_df['date'])

# Capture the email
clean_df['from'] = map(lambda x: filter_email(x), clean_df['from'])
```
